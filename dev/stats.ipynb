```python
import matplotlib.pyplot as plt

def plot_bar(title:str, data:dict):
    labels = ['normal', 'inner_slight', 'inner_moderate', 'inner_severe', 'ball_slight', 'ball_moderate', 'ball_severe', 'outer_slight', 'outer_moderate', 'outer_severe', 'complex']
    colors = ['#fee440', '#90e0ef', '#00b4d8', '#0077b6', '#c7f9cc', '#80ed99', '#57cc99', '#ffb3c6', '#ff8fab', '#fb6f92', '#343a40']

    # 创建柱状图
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # 设置每个子柱状图的宽度
    bar_width = 0.8
    opacity = 0.8
    
    
    # 绘制每个柱状图的子柱
    for i, (key, values) in enumerate(data.items()):
        bottom = 0  # 初始化子柱的底部位置
        for j, value in enumerate(values):
            ax.bar(key, value, bar_width, alpha=opacity, label=labels[j] if i == 0 else '', bottom=bottom, color=colors[j])
            bottom += value  # 更新下一个子柱的底部位置
    
    # 设置图表的标题和标签
    ax.set_title(title)
    ax.legend()
    
    # 显示图表
    plt.tight_layout()
    plt.savefig(f'{title}.pdf')
    plt.show()
```

```python
plot_bar('num distribution',stats_dict_num)
```

```python
plot_bar('time distribution',stats_dict_time)
```
